## Scatter plots by vessel location

```python
# Declare and Initalize variables
N_lad, N_rca, N_lcx, N_lms, N_dx, N_cx, N_om, N_pda, N_int = 0,0,0,0,0,0,0,0,0 #Initalizing variables
Vessel_names = ['LAD', 'RCA', 'LCx', 'LMS', 'Dx', 'Cx', 'OM', 'PDA', 'Int']
V_data = rawdata_array[1:,1]
Vessels_array = list(V_data)
LAD,RCA,LCx,LMS,Dx,Cx,OM,PDA,Int = [],[],[],[],[],[],[],[],[]
Non_LAD,Non_RCA,Non_LCx,Non_LMS,Non_Dx,Non_Cx,Non_OM,Non_PDA,Non_Int = [],[],[],[],[],[],[],[],[]
mr_LAD,mr_RCA,mr_LCx,mr_LMS,mr_Dx,mr_Cx,mr_OM,mr_PDA,mr_Int = y1,y1,y1,y1,y1,y1,y1,y1,y1
d_LAD,d_RCA,d_LCx,d_LMS,d_Dx,d_Cx,d_OM,d_PDA,d_Int = d,d,d,d,d,d,d,d,d

# Format Diameters into 2D array
d_LAD_array = np.array(d_LAD)
d_RCA_array = np.array(d_RCA)
d_LCx_array = np.array(d_LCx)
d_LMS_array = np.array(d_LMS)
d_Dx_array = np.array(d_Dx)
d_Cx_array = np.array(d_Cx)
d_OM_array = np.array(d_OM)
d_PDA_array = np.array(d_PDA)
d_Int_array = np.array(d_Int)

d_LAD_res = d_LAD_array.reshape(-1,3)
d_RCA_res = d_RCA_array.reshape(-1,3)
d_LCx_res = d_LCx_array.reshape(-1,3)
d_LMS_res = d_LMS_array.reshape(-1,3)
d_Dx_res = d_Dx_array.reshape(-1,3)
d_Cx_res = d_Cx_array.reshape(-1,3)
d_OM_res = d_OM_array.reshape(-1,3)
d_PDA_res = d_PDA_array.reshape(-1,3)
d_Int_res = d_Int_array.reshape(-1,3)

# Gather data points for each vessel
xLAD_data = (x1,x2,x3)
xRCA_data = (x1,x2,x3)
xLCx_data = (x1,x2,x3)
xLMS_data = (x1,x2,x3)
xDx_data = (x1,x2,x3)
xCx_data = (x1,x2,x3)
xOM_data = (x1,x2,x3)
xPDA_data = (x1,x2,x3)
xInt_data = (x1,x2,x3)

# Count number of vessels per location
N_lad = Vessels_array.count(Vessel_names[0])
N_rca = Vessels_array.count(Vessel_names[1])
N_lcx = Vessels_array.count(Vessel_names[2])
N_lms = Vessels_array.count(Vessel_names[3])
N_dx = Vessels_array.count(Vessel_names[4])
N_cx = Vessels_array.count(Vessel_names[5])
N_om = Vessels_array.count(Vessel_names[6])
N_pda = Vessels_array.count(Vessel_names[7])
N_int = Vessels_array.count(Vessel_names[8])


# Assign vessels to groups 
for i in range(len(Vessels_array)):
    if(Vessels_array[i] == Vessel_names[0]):
        LAD.append("LAD")
    elif(Vessels_array[i] == Vessel_names[1]):
        RCA.append("RCA")
    elif(Vessels_array[i] == Vessel_names[2]):
        LCx.append("LCx")
    elif(Vessels_array[i] == Vessel_names[3]):
        LMS.append("LMS")
    elif(Vessels_array[i] == Vessel_names[4]):
        Dx.append("Dx")
    elif(Vessels_array[i] == Vessel_names[5]):
        Cx.append("Cx")
    elif(Vessels_array[i] == Vessel_names[6]):
        OM.append("OM")
    elif(Vessels_array[i] == Vessel_names[7]):
        PDA.append("PDA")
    elif(Vessels_array[i] == Vessel_names[8]):
        Int.append("Int")

# Determine unecessary rows for each vessel location
for i in range(len(Vessels_array)):
    if(Vessels_array[i] != Vessel_names[0]):
        Non_LAD.append(i)
    if(Vessels_array[i] != Vessel_names[1]):
        Non_RCA.append(i)
    if(Vessels_array[i] != Vessel_names[2]):
        Non_LCx.append(i)
    if(Vessels_array[i] != Vessel_names[3]):
        Non_LMS.append(i)
    if(Vessels_array[i] != Vessel_names[4]):
        Non_Dx.append(i)
    if(Vessels_array[i] != Vessel_names[5]):
        Non_Cx.append(i)
    if(Vessels_array[i] != Vessel_names[6]):
        Non_OM.append(i)
    if(Vessels_array[i] != Vessel_names[7]):
        Non_PDA.append(i)
    if(Vessels_array[i] != Vessel_names[8]):
        Non_Int.append(i)
```

The deleting section will be done separetly so it is only done once. This is because they are deleted by index,

```python
# Delete unecessary data point found from MR and Diameters, respectively
MR_LAD = np.delete(mr_LAD, [Non_LAD],axis=0)
MR_RCA = np.delete(mr_RCA, [Non_RCA],axis=0)
MR_LCx = np.delete(mr_LCx, [Non_LCx],axis=0)
MR_LMS = np.delete(mr_LMS, [Non_LMS],axis=0)
MR_Dx = np.delete(mr_Dx, [Non_Dx],axis=0)
MR_Cx = np.delete(mr_Cx, [Non_Cx],axis=0)
MR_OM = np.delete(mr_OM, [Non_OM],axis=0)
MR_PDA = np.delete(mr_PDA, [Non_PDA],axis=0)
MR_Int = np.delete(mr_Int, [Non_Int],axis=0)

Dt_LAD = np.delete(d_LAD_res, [Non_LAD],axis=0)
Dt_RCA = np.delete(d_RCA_res, [Non_RCA],axis=0)
Dt_LCx = np.delete(d_LCx_res, [Non_LCx],axis=0)
Dt_LMS = np.delete(d_LMS_res, [Non_LMS],axis=0)
Dt_Dx = np.delete(d_Dx_res, [Non_Dx],axis=0)
Dt_Cx = np.delete(d_Cx_res, [Non_Cx],axis=0)
Dt_OM = np.delete(d_OM_res, [Non_OM],axis=0)
Dt_PDA = np.delete(d_PDA_res, [Non_PDA],axis=0)
Dt_Int = np.delete(d_Int_res, [Non_Int],axis=0)

# Format diameter data
D_LAD = np.array(Dt_LAD)
D_RCA = np.array(Dt_RCA)
D_LCx = np.array(Dt_LCx)
D_LMS = np.array(Dt_LMS)
D_Dx = np.array(Dt_Dx)
D_Cx = np.array(Dt_Cx)
D_OM = np.array(Dt_OM)
D_PDA = np.array(Dt_PDA)
D_Int = np.array(Dt_Int)
```

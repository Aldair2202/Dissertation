### Bland-Altman Plot

```python
 def bland_altman_plot(data1, data2, *args, **kwargs):
    data1     = np.asarray(data1)
    data2     = np.asarray(data2)
    mean      = np.mean([data1, data2], axis=0)
    diff      = data1 - data2                   # Difference between data1 and data2
    md        = np.mean(diff)                   # Mean of the difference
    sd        = np.std(diff, axis=0)            # Standard deviation of the difference

    plt.scatter(mean, diff, *args, **kwargs)
    plt.axhline(md,           color='gray', linestyle='--')
    plt.axhline(md + 1.96*sd, color='gray', linestyle='--')
    plt.axhline(md - 1.96*sd, color='gray', linestyle='--')

plt.figure(0)
bland_altman_plot(MR_filtered.astype('float'),ModelEq_re, color='blue')
plt.title('Bland-Altman Plot')
plt.xlabel('Average of Actual Myocardial resistance and Predicted Myocardial Resistance')
plt.ylabel('Difference between Actual Myocardial resistance and Predicted Myocardial Resistance')
```

### Correlation plots

Here, the Myocardial resistance is being scaled as its log10 value.

```python
def getMRlog(data):
    data_array = np.array(data)
    data_vector = data_array.reshape(-1,1)
    MR_log10 = np.log10(data_vector.astype('float'))
    
    return MR_log10

MRL = getMRlog(MR_filtered)
plt.scatter(DIn_filtered,MRL,label="Inlet")
plt.scatter(DOut_filtered,MRL,label="Outlet")
plt.scatter(DMin_filtered,MRL,label="Minimum")
plt.scatter(Age_filtered,MRL,label="Age")
plt.scatter(Height_filtered,MRL,label="Height")
plt.scatter(Weight_filtered,MRL,label="Weight")
plt.scatter(Heart_rate_filtered,MRL,label="Heart rate")
plt.xlabel('Parameters (see lenged)')
plt.ylabel('LOG Myocardial resistance')
plt.title('MR vs Inlet Diameter')
#plt.legend(loc=8)
```
